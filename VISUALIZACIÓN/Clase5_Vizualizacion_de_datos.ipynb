###SUBPLOTS
https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig,(ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig,axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x, np.cos(x),'y')
axes[0,1].plot(x, np.sin(x),'b')
axes[0,2].plot(x, np.tan(x),'r')
axes[0,3].plot(x, np.tan(x)**2,'g')
```

```python
fig,((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2, ncols=4)
ax1.plot(x, np.cos(x),'y')
ax2.plot(x, np.sin(x),'b')
ax3.plot(x, np.tan(x),'r')
ax4.plot(x, np.tan(x)**2,'g')
ax5.plot(x, np.cos(x),'y')
ax6.plot(x, np.sin(x),'b')
ax7.plot(x, np.tan(x),'r')
ax8.plot(x, np.tan(x)**2,'g')
fig.tight_layout()
```

###LEYENDAS, ETIQUETAS, TÍTULOS, TAMAÑOS
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize = (5,5))
axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Relación X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relación Y- X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y, label='sin(x)')
plt.title('Este es un título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

### RELATION
https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.head(2)
```

```python
plt.figure(figsize= (8,8))
sns.set(style="ticks", palette=None, font_scale=1.3)
markers = {'Lunch': 'D', 'Dinner':'s'}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
plt.figure(figsize= (8,8))
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
# Aquí implementamos la ralación
plt.figure(figsize= (10,10))
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time')
plt.show()
```

###JOINPLOT / PAIRPLOT
