```python
import pandas as pd
import numpy as np
```

```python
dic={'autos':['VW','Nissan','Ford'],'color':['rojo','azul','negro']}
autos_posibles=['VW','Nissan','Ford','Renault', 'Toyota','chevrolet']
colores_posibles=['rojo','azul','negro','amarillo','plata', 'blanco']
```

```python
A=[]
C=[]
for i in range(500):
    A.append(np.random.choice(autos_posibles))
    C.append(np.random.choice(colores_posibles))
```

```python
dic2={'autos':A,'color':C}
```

```python
pd.DataFrame(dic2)
```

```python
DataCoches=pd.DataFrame(dic2)
```

```python
#HACER LLAMADOS DE LOS COCHES DE LA COLUMNA autos
DataCoches['autos']
```

```python
# hacer llamados de todos los colores posibles de DataCoches
DataCoches['color']
```

```python
v=[]
for i in range(500):
    v.append(np.random.choice(range(10,121)))
```

```python
DataCoches['velocidad']=v
```

```python
DataCoches
```

```python
#agrupando grupo de autos con colores y sus velocidades y sacando su media 
DataCoches.groupby('color')['velocidad'].mean()
```

```python
DataCoches.groupby('color')['velocidad'].sum()
```

```python
DataCoches.groupby('color')['velocidad'].count()
```

```python
DataCoches.groupby('color')['velocidad'].describe()
```

```python
DataCoches.pivot_table(index='autos',columns='color',aggfunc='count',margins=True,margins_name='Total')
```

```python
DataCoches.pivot_table(index='autos',columns='color',aggfunc='count',margins=True,margins_name='Total').to_csv('tablasCoches.csv')
```

```python
DataCoches.pivot_table(index='autos',columns='color',aggfunc='count',margins=True,margins_name='Total').to_excel('tablasCoches.xlsx')
```
