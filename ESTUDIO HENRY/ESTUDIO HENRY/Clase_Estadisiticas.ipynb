```python
def obtenerMediana(arreglo):

    numeros = [numero for fila in arreglo for numero in fila]
    numeros.sort() # ordena arreglo
    mediana = numeros[0]
    tamano = len(numeros)

    if( tamano % 2 != 0):
      mediana = numeros[(tamano // 2)]
    else:    
      numero_2 = numeros[(tamano // 2) - 1]
      numero_1 = numeros[(tamano // 2)]
      mediana = ((numero_1 + numero_2) / 2)

    return mediana

obtenerMediana(muestra)   
```

```python
#VARIANZA
def obtenerVarianza(arreglo):
  numeros = [numero for fila in arreglo for numero in fila]
  suma , promedio, numero_i, x, y = 0, 0, 0, 0, 0
  tamano = len(numeros)

  for i in numeros:
    suma += i
    promedio = suma / tamano
  
  for i in numeros:
    numero_i = i - promedio
    x = numero_i * numero_i
    y += x
  return y / tamano

obtenerVarianza(muestra)
  
```

```python
#DESVIACIÓN ESTANDAR

def obtenerDesviacionStandar(arreglo):
  numeros = [numero for fila in arreglo for numero in fila]
  suma , promedio, numero_i, x, y = 0, 0, 0, 0, 0
  tamano = len(numeros)

  for i in numeros:
    suma += i
    promedio = suma / tamano
  
  for i in numeros:
    numero_i = i - promedio
    x = numero_i * numero_i
    y += x
  return (y / (tamano ))**0.5

obtenerDesviacionStandar(muestra)
```
