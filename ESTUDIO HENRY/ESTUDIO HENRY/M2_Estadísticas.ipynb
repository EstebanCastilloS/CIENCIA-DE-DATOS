```python
def obtenerMediana(arreglo):
   # numeros = []
   # for fila in arreglo:
   #    for numero in fila:
   #       numeros.append(numero)
   numeros = [numero for fila in arreglo for numero in fila]
   numeros.sort()
   tamano = len(numeros)
   mediana = numeros[0]

   if (tamano % 2 > 0): 
     # cuando es par la lista
     mediana = numeros[tamano // 2]
   else:
     numero_1 = numeros[tamano // 2]
     numero_2 = numeros[tamano // 2 - 1]
     mediana = ((numero_1 + numero_2) / 2)

   return mediana  

print(obtenerMediana(muestra))
```

### MODA

```python
#Moda
#print(stats.mode(muestra.flatten())[0])

def obtenerModa(arreglo):
   numeros = [numero for fila in arreglo for numero in fila] #Paso la matriz a una lista
   unicos = list(set(numeros)) #Obtengo valores únicos
   moda, maximo = 0,0
   for indice, numero in enumerate(unicos):
      conteo = numeros.count(numero) #Busco la mayor repetición de ese valor 
      if conteo > maximo:
         maximo = conteo
         moda = unicos[indice]
   return moda

obtenerModa(muestra)
```

### VARIANZA

```python
#Varianza
print(np.var(muestra))
```

```python
# VARIANZA POBLACIONAL
def obtenerVarianzaPoblacional(arreglo):
   numeros = [numero for fila in arreglo for numero in fila] #Paso la matriz a una lista
   suma, promedio, numero_j, x, y = 0, 0, 0, 0, 0
   tamano = len(numeros)

   for i in numeros:
     suma += i
     promedio = suma / tamano
   for j in numeros:
     numero_j = j - promedio
     x = numero_j * numero_j 
     y += x    
   return (y / tamano)   
   
obtenerVarianzaPoblacional(muestra)
```

```python
# VARIENZA MUESTRAL
def obtenerVarianzaMuestral(arreglo):
   numeros = [numero for fila in arreglo for numero in fila] #Paso la matriz a una lista
   unicos = list(set(numeros)) #Obtengo valores únicos
   suma, promedio, numero_j, x, y = 0, 0, 0, 0, 0
   tamano = len(numeros)
   for i in numeros:
     suma += i
     promedio = suma / tamano
   for j in numeros:
     numero_j -= promedio
     x = numero_j * numero_j 
     y += x    
   return y / (tamano - 1)  
   
obtenerVarianzaMuestral(muestra)
```

```python
# VARIENZA MUESTRAL
def obtenerVarianzaDesviacionEstandar(arreglo):
   numeros = [numero for fila in arreglo for numero in fila] #Paso la matriz a una lista
   unicos = list(set(numeros)) #Obtengo valores únicos
   suma, promedio, numero_j, x, y = 0, 0, 0, 0, 0
   tamano = len(numeros)
   for i in numeros:
     suma += i
     promedio = suma / tamano
   for j in numeros:
     numero_j = j - promedio
     x = numero_j * numero_j 
     y += x    
   return (y / (tamano - 1))**0.5 
   
obtenerVarianzaDesviacionEstandar(muestra)
```
