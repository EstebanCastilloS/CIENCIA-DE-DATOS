### 1.1 DISTRIBUCIONES DE FRECUENCIA
_______________________________________________________________

Distribuciones de frecuencia para una variable cualitativa

```python
frecuencia = datos['Sexo'].value_counts()
frecuencia
```

```python
#SACANDO LOS DATOS DE MANERA PORCENTUAL
porcentual = datos['Sexo'].value_counts(normalize = True)*100
porcentual
```

```python
distribucion_frecuencias_cualitativas =pd.DataFrame({'Frecuencia': frecuencia, 'Porcentual (%)':porcentual})
distribucion_frecuencias_cualitativas
```
