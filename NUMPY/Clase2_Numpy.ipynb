##AGREGAR O ELIMINAR DIMENSIONES

```python
#Agregando cantidades de dimensiones
vector10 = np.array([1,2,3,4], ndmin=10)
print(vector10)
vector10.ndim
```

```python
expand = np.expand_dims(np.array([1,2,3]), axis=0)
print(expand)
expand.ndim
```

##ELIMINAR DIMENSIONES

```python
# QUITAR LAS DIMENSIONES QUE NO SE ESTÉN UTILIZANDO

print(vector10,vector10.ndim)
vector_2 = np.squeeze(vector10)
print(vector_2, vector_2.ndim)
```

```python
a = np.array([[1,2],[3,4]])
print(a)
b = np.array([5,6])
print(b)
```

```python
# Ahora vamos a concatenar 2 arrays
# sabemos que a y b son de dimensiones diferentes hay que acomodar a "b"

b = np.expand_dims(b,axis=0)
b.ndim
```

```python
# ahora ya se tiene las mismas dimensiones y procedemos a concatenar de manera de columnas
np.concatenate((a,b), axis=0)
```

```python
#Ahora vamos a concatenar de manera de filas
np.concatenate((a,b.T), axis=1)
```
